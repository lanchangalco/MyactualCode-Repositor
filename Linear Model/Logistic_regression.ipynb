```python
# REF_NO,post_code , post_area  : drop 
# children : Zero : 0 , 4+ : 4 and then convert to numeric
# age_band : dummies 
# status , occupation , occupation_partner , home_status,family_income : dummies
# self_employed, ` : dummies
# TVArea , Region , gender : dummies
# Revenue Grid : 1,2 : 1,0
```

```python
bd.drop(['REF_NO','post_code','post_area'],axis=1,inplace=True)
```

```python
#bd.loc[bd["children"]=="Zero","children"]="0"
#bd.loc[bd["children"]=="4+","children"]="4"
#bd["children"]=pd.to_numeric(bd["children"],errors="coerce")
bd['children'].value_counts()
```

```python
bd['children']=np.where(bd['children']=='Zero','0',bd['children'])
bd['children']=np.where(bd['children']=='4+',4,bd['children'])
bd['children']=pd.to_numeric(bd['children'],errors='coerce')
```

```python
bd['family_income'].value_counts()
```

```python
k=bd['family_income'].str.split('>=',expand=True)
```

```python
k
```

```python
k[0]=k[0].str.replace('<','')
k[1]=k[1].str.replace('<','')
k[0]=k[0].str.replace(',','')
k[1]=k[1].str.replace(',','')
```

```python
k
```

```python
#where 1st column empty take from 2nd
#where ever its none also

k.loc[k[0]=='',0]=k[1]
k.loc[k[1].isnull(),1]=k[0]
```

```python
k
```

```python
k[0]=pd.to_numeric(k[0],errors='coerce')
k[1]=pd.to_numeric(k[1],errors='coerce')

bd['fi']=0.5*(k[0]+k[1])
del bd['family_income']
```

```python
bd['Revenue Grid'].value_counts()
```

```python
bd['Revenue Grid']=(bd['Revenue Grid']==1).astype(int)
```

```python
cat_var=bd.select_dtypes(['object']).columns
```

```python
cat_var
```

```python
for col in cat_var:
    dummy=pd.get_dummies(bd[col],drop_first=True,prefix=col)
    bd=pd.concat([bd,dummy],axis=1)
    del bd[col]
del dummy
```

```python
bd.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
bd_train,bd_test=train_test_split(bd,test_size=0.2,random_state=2)
bd_train.isnull().sum()
```

```python
bd_test.isnull().sum()
```

```python
for col in ['children','fi']:
    bd_train.loc[bd_train[col].isnull(),col]=bd_train[col].mean()
    bd_test.loc[bd_test[col].isnull(),col]=bd_train[col].mean()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
logr=LogisticRegression(class_weight='balanced')
```

```python
x_train=bd_train.drop('Revenue Grid',axis=1)
y_train=bd_train['Revenue Grid']
x_test=bd_test.drop('Revenue Grid',axis=1)
y_test=bd_test['Revenue Grid']
```

```python
logr.fit(x_train,y_train)
```

```python
logr.predict(x_test)
```

```python
logr.predict_proba(x_test)
```

```python
logr.classes_
```

```python
predicted_probs=logr.predict_proba(x_test)[:,1]
```

```python
predicted_probs
```

```python
roc_auc_score(y_test,predicted_probs)
```

```python
cutoffs=np.linspace(0.001,0.99,99)
cutoffs
```

```python
train_score=logr.predict_proba(x_train)[:,1]
real=y_train
```

```python
KS_all=[]
for cutoff in cutoffs:
    predicted=(train_score>cutoff).astype(int)
    
    TP=((predicted==1) & (real==1)).sum()
    TN=((predicted==0) & (real==0)).sum()
    FP=((predicted==1) & (real==0)).sum()
    FN=((predicted==0) & (real==1)).sum()
    
    P=TP+FN
    N=TN+FP
    
    KS=(TP/P)-(FP/N)
    
    KS_all.append(KS)
    
# do this for fb
    
```

```python
mycutoff=cutoffs[KS_all==max(KS_all)]
mycutoff
```

```python
logr.intercept_
```

```python
list(zip(x_train.columns,logr.coef_[0]))
```

```python
test_score=logr.predict_proba(x_test)[:,1]
```
