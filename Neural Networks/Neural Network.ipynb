```python
catcols=cd.select_dtypes(['object']).columns
```

```python
for col in catcols:
    freqs=cd[col].value_counts()
    k=freqs.index[freqs>99][:-1]
    for cat in k:
        name=col+'_'+cat
        cd[name]=(cd[col]==cat).astype(int)
    del cd[col]
    print(col)
```

```python
cd.shape
```

```python
cd_train,cd_test=train_test_split(cd,test_size=0.2,random_state=2)
```

```python
cd_train.reset_index(drop=True,inplace=True)
cd_test.reset_index(drop=True,inplace=True)
```

```python
x_train=cd_train.drop(["Y"],axis=1)
y_train=cd_train['Y']

x_test=cd_test.drop(["Y"],axis=1)
y_test=cd_test['Y']
```

```python
x_train.shape
```

```python
hidden_layer_sizes=[(5,10,5),(20,10),(10,20),(15,11),(20,5),(5,10)]

kf=KFold(n_splits=10)

auc_list=[]
i=0

for a in hidden_layer_sizes:
    print('hidden layer input and sequence no: ',a,'&',i+1)
    print('\n')
    i+=1
    nn=MLPClassifier(hidden_layer_sizes=a,activation='relu',learning_rate_init=0.01)
    error=0
    j=0
    for train,test in kf.split(x_train):
        print('fold number :',j+1)
        j+=1
        nn.fit(x_train.loc[train],y_train[train])
        p=nn.predict_proba(x_train.loc[test])[:,1]
        error+=roc_auc_score(y_train[test],p)
    auc_list.append(error/10)
    print(a,' : ',error/10 )
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(5,20),activation='relu',learning_rate_init=.01)
```

```python
mlp.fit(x_train,y_train)
```

```python
roc_auc_score(y_train,mlp.predict_proba(x_train)[:,1])
```

```python
roc_auc_score(y_test,mlp.predict_proba(x_test)[:,1])
```
